```python
# Jetzt bauen wir das in den Loop ein: 

seite_eins = []

for element in tr_list:
    
    #wichtig: hier wird die td_list bei jedem loop mit dem nächsten ELement neu zusammengesetzt, 
    #damit wir im folgenden die Daten rausziehen können:
    td_list = element.find_all("td")

    
    url = td_list[0].find("a")["href"]
    geschäft = td_list[1].text
    stand = td_list[2].text
    zuständig = td_list[3].text
    geschäftsart = td_list[4].text
    status = td_list[5].text

    mini_dict = {"URL": url,
             "Geschäft": geschäft,
             "Stand": stand,
             "zuständig": zuständig, 
             "Geschäftsart": geschäftsart, 
             "Status": status}
    
    seite_eins.append(mini_dict)    
```

```python
# Jetzt bauen wir das in einen übergeordneten Loop ein, der jeweils die nächte Seite aufruft, 
# indem er am Ende der URL die Zahl ändert.

import requests
from bs4 import BeautifulSoup
import pandas as pd

alle_vorstösse = []

website_anfang = "https://kr-geschaefte.zug.ch/gast/geschaefte?commit=Filtern&geschaeft_filter%5Babgeschlossen_bis%5D=&geschaeft_filter%5Babgeschlossen_von%5D=&geschaeft_filter%5Barten_refs%5D%5B%5D=&geschaeft_filter%5Beingereicht_bis%5D=&geschaeft_filter%5Beingereicht_von%5D=&geschaeft_filter%5Bfrist_bis%5D=&geschaeft_filter%5Bhistorische_staende_refs%5D%5B%5D=&geschaeft_filter%5Bkommissionen_refs%5D%5B%5D=&geschaeft_filter%5Bstaende_refs%5D%5B%5D=&geschaeft_filter%5Bstatus_ids%5D%5B%5D=haengig&geschaeft_filter%5Bstatus_ids%5D%5B%5D=abgeschlossen&geschaeft_filter%5Bstatus_ids%5D%5B%5D=&geschaeft_filter%5Btitel%5D=&geschaeft_filter%5Bzustaendig_refs%5D%5B%5D=&page=" 

for page in range (1,41):
    r = requests.get(website_anfang + str(page))
    soup = BeautifulSoup(r.text, "html.parser")
    tr_list = soup.find_all("tr")
    tr_list = soup.find_all("tr")[2:]
        
    for element in tr_list:

        td_list = element.find_all("td")
    
        url = td_list[0].find("a")["href"]
        geschäft = td_list[1].text
        stand = td_list[2].text
        zuständig = td_list[3].text
        geschäftsart = td_list[4].text
        status = td_list[5].text

        mini_dict = {"URL": url,
                     "Geschäft": geschäft,
                     "Stand": stand,
                     "zuständig": zuständig,
                     "Geschäftsart": geschäftsart, 
                     "Status": status}
    
        alle_vorstösse.append(mini_dict)

pd.DataFrame(alle_vorstösse)
```

#### Jetzt versuchen wir die kompletten URL zu generieren:

```python
url_start = "https://kr-geschaefte.zug.ch"
url_liste = []

for eintrag in alle_vorstösse:
    komplett_url = url_start + str(eintrag["URL"])
    geschäft = eintrag["Geschäft"] 
    
    mini_dict = {"Geschäft": geschäft, "URL": komplett_url}

    url_liste.append(mini_dict)

url_df = pd.DataFrame(url_liste)
url_df
```
