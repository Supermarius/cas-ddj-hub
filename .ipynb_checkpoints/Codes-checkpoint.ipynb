### Hundenamen-Suche

```python
###Eingabemaske
search = input("Gib hier den gesuchten Namen ein: ")
year1 = input("Gib hier das Startjahr ein: ")
year2 = input("Gib hier das Endjahr ein: ")

year1 = int(year1)
year2 = int(year2)

###Rauslesen und Ausgabe
with open("20151001_hundenamen.csv") as file:
    for line in file:
        data = line.split(",")
        
        name = data[0]
        year = data[1]
        
        if search not in name:
            continue                           
        elif search in name and int(year) >= year1 and int(year) <= year2:
            print(data)
```

### Funktion Namenmixer

```python
def Namenmixer(name):
    
    ###Vorbereitungen:
    import random
    buchstabenmax=1
    zerlegt = []
    mix = [] 

    ###Namen zerlegen in Liste:
    for letter in name:
        zerlegt.append(letter)
    
    ###Namen zufällig neu zusamensetzen als Mix:   
    for letter in name:
        index = random.randint(0, len(name)-buchstabenmax)
        genommen = zerlegt.pop(index)
        ## print(genommen)
        mix.append(genommen)
        buchstabenmax = buchstabenmax +1

    ###Drucke neuen Namen:  
    newname = ("".join(mix))
    print("Hey " + name + ", dein frisch gemixter Name lautet: " + newname.capitalize() +".")    
```

```python
### Anwendung: 
Namenmixer("Marius")
```

```python
def Multi_Mixer(name):
    for case in name: 
        Namenmixer(name)

Multi_Mixer("Marius")
```
