```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
r = requests.get("https://www.gr.ch/DE/institutionen/parlament/protokolle/Seiten/IndexProtokolleGR.aspx")
```

```python
content = r.text
```

```python
soup = BeautifulSoup(content, "html.parser")
```

```python
soup
```

```python
linkliste = soup.find_all('div')
```

```python
#in diesem unter_div hier sind die links drin (erneut alle durchprobiert): 
linkliste[5].find_all("div")[33]
```

```python
#neuer Anlauf, EINFACHER dank Ansteuern mit Entwicklertool:
linkliste = soup.find("ul",{"class": "TitleOnly"}).find_all("a")
```

```python
linkliste
```

```python
#Regex importieren, um URL rauszubekommen
import re
```

```python
#zunächst mal eine einzelne:
eintrag1 = str(linkliste[0])
url = re.search(r"https://www\..*aspx", eintrag1, re.IGNORECASE)
url.group()
```

```python
#jetzt ein Loop und der alle Ergebnisse in einer Liste ablegt:
saubere_linkliste = []

for nummer in range(0, len(linkliste)):
    eintrag1 = str(linkliste[nummer])
    url = re.search(r"https://www\..*aspx", eintrag1, re.IGNORECASE)
    url = url.group()
    saubere_linkliste.append(url)

saubere_linkliste
```

```python
#jetzt mal eins ansteuern und sublinks rauszwacken:
website = requests.get("https://www.gr.ch/DE/institutionen/parlament/protokolle/Seiten/Protokolle2010.aspx")
website_lesbar = website.text
website_suppe = BeautifulSoup(website_lesbar, "html.parser")
sub_links = website_suppe.find("ul",{"class": "Links"}).find_all("a")
sub_links
```

```python
#jetzt jede dieser Seiten aufrufen und dort Linkliste zueinzelnen Sitzungen rausholen ("<ul class="Links CQWP">")

saubere_linkliste2 = []

for link in saubere_linkliste:
    website = requests.get(link)
    website_lesbar = website.text
    website_suppe = BeautifulSoup(website_lesbar, "html.parser")
    sub_links = website_suppe.find("ul",{"class": "Links"}).find_all("a")
        
    for nummer in range(0, len(sub_links)):
        eintrag1 = str(sub_links[nummer])
        #print(eintrag1) #zur Kontrolle
        #input("Press Enter to continue...") #zur Kontrolle
        url = re.search(r"https://www\..*htm", eintrag1, re.IGNORECASE) # Achtung: her ändert sich der URL-Code zu htm
        url = url.group()
        saubere_linkliste2.append(url)

saubere_linkliste2

# Funktioniert nicht, weil irgendwo auf einer der Jahreswebsites der class-Tag für die Links ändert!
```
