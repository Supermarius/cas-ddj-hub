### Hundenamen-Counter Version 2

```python
### Hier definiere ich meine Variablen und Listen:

countedname = "xxxxxx"
computed_element = "xxxxxx"

listoflists = []
namelist = []
countlist = []
countlist_clean = []

### Hier öffne ich das File und zerlege die Zeilen in Listen, die ich in einer "List of Lists" sammle:

with open("20151001_hundenamen.csv") as file:
    for line in file:
        data = line.split(",")
        listoflists.append(data)
        
### Hier mache ich einen Dictionary, in der die Namen mit der Anzahl Nennungen kombiniert
### (Achtung: das funktioniert so nur, wenn die Namen in der Liste schon alphabetisch geordent sind, 
### weil das Programm jeweils nut mit der vorangegangenen Zeile vergleicht... (Man müste also zuerst Liste sortieren!)
        
for data in listoflists:
    name = data[0]
    namelist.append(name) 
    
for name in namelist:
    if name == countedname:
        continue
    else:
        counter = namelist.count(name)
        dictionary = {"Name" : str(name), "Anzahl": str(counter)}
        countlist.append(dictionary)

### Hier lösche ich die Zeilen raus, die das gleiche beinhalten wie die vorangegangene > bereinigte Liste

for element in countlist:
    if element != computed_element:
        computed_element = element
        countlist_clean.append(element)

### Hier sortiere ich die bereinigte Liste nach der Zahl:
        
def get_Anzahl(item):
    return item["Anzahl"]        
   
countlist_clean.sort(key=get_Anzahl, reverse=True)

for element in countlist_clean:
    print(element)
```

### Funktion Namenmixer

```python
def Namenmixer(name):
    
    ###Vorbereitungen:
    import random
    buchstabenmax=1
    zerlegt = []
    mix = [] 

    ###Namen zerlegen in Liste:
    for letter in name:
        zerlegt.append(letter)
    
    ###Namen zufällig neu zusamensetzen als Mix:   
    for letter in name:
        index = random.randint(0, len(name)-buchstabenmax)
        genommen = zerlegt.pop(index)
        ## print(genommen)
        mix.append(genommen)
        buchstabenmax = buchstabenmax +1

    ###Drucke neuen Namen:  
    newname = ("".join(mix))
    print("Hey " + name + ", dein frisch gemixter Name lautet: " + newname.capitalize() +".")    
```

```python
### Anwendung: 
Namenmixer("Marius")
```
