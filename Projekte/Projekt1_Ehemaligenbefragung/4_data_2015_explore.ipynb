So zupfen wir den Namen raus:

Jetzt können wir das also in unsere loop einbauen:

```python
allenamen = []

for file in file_namen:
    regex = r".*\b(?= Anhang)"
    name = re.search(regex, file, re.IGNORECASE).group()
    
    allenamen.append(name)
    
    print(name)
```

Das hier ist das Gleiche für die Listen von 2018:

```python
file_namen2 = os.listdir('Lieferung_SEB-Schulergebnisse_2/Schulberichte/Tabellen_2018')
```

```python
for file in file_namen2:
    regex = r"(?<=4 ).*\b(?= Anhang)"  # leicht angepasst, weil hier die Files mit einer 4 beginnen
    name = re.search(regex, file, re.IGNORECASE).group()
    
    allenamen.append(name)
```

Schön wäre, wen wir die KS noch wegbringen - und die Gym und Gymnasium am Schluss:

```python
allenamen = []

for file in file_namen:
    regex = r".*\b(?= Anhang)"
    name = re.search(regex, file, re.IGNORECASE).group()
    if "Gym" in name:
        regex = r".*\b(?= Gym.*)"
        name = re.search(regex, name, re.IGNORECASE).group()
    if "KS" in name:
        regex = r"(?<=KS ).*\b"
        name = re.search(regex, name, re.IGNORECASE).group()
    allenamen.append(name)
    
    print(name)
```

Also, nochmals von vorn:

```python
jahr = 2015
dic_list = []

for file in file_namen:
    regex = r".*\b(?= Anhang)"
    name = re.search(regex, file, re.IGNORECASE).group()
    if "Gym" in name:
        regex = r".*\b(?= Gym.*)"
        name = re.search(regex, name, re.IGNORECASE).group()
    if "KS" in name:
        regex = r"(?<=KS ).*\b"
        name = re.search(regex, name, re.IGNORECASE).group()

    # Jetzt öffnen wir das jeweilige File als Dataframe:
    
    df = pd.read_excel("Lieferung_SEB-Schulergebnisse_2/Schulberichte/Tabellen_2015/" + str(file))
    
    zufr = df.loc[20,"Unnamed: 3"]
    vorb_f = df.loc[22,"Unnamed: 3"]
    vorb_p = df.loc[24,"Unnamed: 3"]
    prob = df.loc[56,"Unnamed: 3"]
    urt = df.loc[58,"Unnamed: 3"]
    zus = df.loc[60,"Unnamed: 3"]
    konf = df.loc[62,"Unnamed: 3"]

    mini_dict = {"jahr": jahr, 
                 "name": name, 
                 "zufr": zufr, 
                 "vorb_f": vorb_f, 
                 "vorb_p":vorb_p, 
                 "prob": prob, 
                 "urt": urt, 
                 "zus": zus, 
                 "konf": konf}
    dic_list.append(mini_dict)    
```

```python
df_final = pd.DataFrame(dic_list)
df_final.to_csv("Pandas_Results2015.csv")
df_final
```
