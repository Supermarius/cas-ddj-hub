```python
#Prozentualer Anteil
df["Festsetzung"].value_counts(normalize=True)
```

```python
df_Anteil = pd.DataFrame((df["Festsetzung"].value_counts(normalize=True)*100).round(3))
pd.set_option('display.max_rows', df.shape[0])
df_Anteil

#Die grosen Festsetzungen: 43 Prozent wurden am §1979 festgesetzt, 19% 2018 und 
```

```python
df_Anteil.plot()
```

```python
#Gruppiert nach Jahren:
df_Anteil.groupby(df_Anteil.index.year)["Festsetzung"].sum().plot(figsize=(8,4), kind='bar')
```
