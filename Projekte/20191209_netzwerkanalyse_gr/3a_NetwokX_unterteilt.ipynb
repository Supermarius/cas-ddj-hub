```python
import warnings
warnings.filterwarnings('ignore')

import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd

file = open("Dataframe_bereinigt_blockuebergreifend.csv", mode="r")
df2= pd.read_csv(file)
df2 = df2.set_index("Unnamed: 0")
```

Problem bei Datierung: Ein paar Geschäfte (die mit anderen verknüpft sind) haben ein fehlerhaftes Datum ("08.01.20"), das die to_datetime-Kovertierung stört. Diese erseten wird mit dem Wert 31.12. plus dem entsprechenden Wert aus dem Geschäft. Das machen wir so:

```python
fehlerposition3 = df2[df2["Datum"] == "08.01.20"].index.tolist()

for zeile in fehlerposition3:
    Jahr = df2.loc[zeile, "Geschäft"].split("/")[0]
    Ersatzdatum = "31.12." + Jahr
    df2.loc[zeile, "Datum"] = df2.loc[zeile, "Datum"].replace("08.01.20", Ersatzdatum)
```

```python
df2.loc[110, "Datum"] = df2.loc[zeile, "Datum"].replace("30.6.", "30.6.2004")
df2.loc[64, "Datum"] = df2.loc[zeile, "Datum"].replace("29.01.20", "28.08.2002")
df2.loc[305, "Datum"] = df2.loc[zeile, "Datum"].replace("29.01.20", "09.06.2004")
df2.loc[308, "Datum"] = df2.loc[zeile, "Datum"].replace("29.01.20", "30.5.2001")
df2.loc[416, "Datum"] = df2.loc[zeile, "Datum"].replace("29.01.20", "24.03.2004")
```

### Netzwerk-Export für Gephi

Weil wirs nicht geschafft haben, in Gephi dynamische Graphs mit Timestamps zu machen, bauen wir jetzt einen Loop, der uns für jede Legislatur ein separates gefx-File baut:

```python
for frame in dataframe_list:
    #müssen neuen Index machen, sonst kann man die einzelnen locs nicht ansteuern...
    frame.reset_index(inplace = True)
    Netz = nx.Graph()

    for a in range(0,(len(frame)-1)):
        Person1 = str(frame.loc[a]["Einreichender 1"])
        Person2 = str(frame.loc[a]["Einreichender 2"])
        Partei1 = str(frame.loc[a]["Partei 1"])
        Partei2 = str(frame.loc[a]["Partei 2"])
        Netz.add_node(Person1, Partei = Partei1)
        Netz.add_node(Person2, Partei = Partei2)

    #Als zweites machen wir die edges und versehen sie mit eindem datetime-Datum:
    for a in range(0,(len(frame)-1)):
        Person1 = str(frame.loc[a]["Einreichender 1"])
        Person2 = str(frame.loc[a]["Einreichender 2"])
        date = str(frame.loc[a]["Datum"])
        Departement = str(frame.loc[a]["Departement"])
    
        #Beziehungen, die mehrmals vorkommen, sollte gewichtet werden (weight +1)
        if Netz.has_edge(Person1, Person2):
            Netz[Person1][Person2]["weight"] += 1
        else:
            Netz.add_edge(Person1, Person2, datum = date, thema = Departement, weight=1)
    
    nx.write_gexf(Netz, "Paare_"+str(frame.loc[0]["Geschäft"].split("/")[0])+".gexf")
```

Und jetzt zur Analyse: 

```python
#generelle Infos zum aktuellen Netz
print(nx.info(Netz))
```

**degree** = Anzahl der Kanten die mit dem Knoten verbunden sind<br>
**gewichteter degree** = summe der gewichteten Kanten

```python
wdegree = dict(Netz.degree(weight='weight'))
#wdegree

#So bekommen wir de Leader zu diesem Wert
max(Netz, key=wdegree.get)
```

**degree centrality** = Wieviele Kanten hat ein Knoten prozentual gemessen am möglichen Maxmimum?<br>
(gemessen an diesem Netz von Leuten, die über partei- oder blockgrenzen miteinander im Austausch stehen.)

```python
degc = nx.degree_centrality(Netz)
#sorted(degc.items(), key=itemgetter(1), reverse=True)
```

**betweenness centrality** =  Summe der Anteile aller kürzesten Pfade von x nach y, die durch diesen Knoten laufen

```python
from operator import itemgetter

#Knoten die mehrere Personen verbinden, oder oft dazwischen liegen haben eine hohe Betweenness.
bet = nx.betweenness_centrality(Netz)
#sorted(bet.items(), key=itemgetter(1), reverse=True)
```

**closeness centrality** = summe der kürzesten Pfade vom Knoten zu allen anderen Knoten 

```python
clos = nx.closeness_centrality(Netz)
#sorted(clos.items(), key=itemgetter(1), reverse=True)
```

**eigenvector centrality** = gewichtet Knoten basierend auf der Idee, dass Knoten mit hohen Gewichten mehr zur Zentrelität beitragen als Knoten mit niedrigen Gewichten. (Wer mit wichtigen Leuten verbunden ist, ist selbst wichtiger)

```python
eig = nx.eigenvector_centrality(Netz)
#sorted(eig.items(), key=itemgetter(1), reverse=True)
```

Und hier die Übersicht:

```python
centrality_measures = {
    'degree': degc,
    'betweenness': bet,
    'closeness': clos,
    #"eigenvector": eig,
}
centrality = pd.DataFrame(centrality_measures)
centrality.sort_values("betweenness", ascending = False).head(10)
```
