```python
hundedict = {}

with open("20151001_hundenamen.csv") as file:
    for line in file:
        ### hier werden die Sonderzeichen rausgelöscht
        line = line.replace('"',"")
        line = line.replace('\n',"")
        ### hier wird die Zeile in eine liste konvertiert
        hundedata = line.split(",")
        name = hundedata[0]
        if name not in hundedict:
            hundedict[name] = 1
        else:
            hundedict[name] = hundedict[name] +1
            
def mysort(item):
    return item[1]

for key,value in sorted(hundedict.items(), key=mysort, reverse=True):
    print("Name '%s' |  Anzahl %s "  % (key,value))
```
