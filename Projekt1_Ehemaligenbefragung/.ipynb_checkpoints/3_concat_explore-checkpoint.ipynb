```python
import pandas as pd
```

```python
data2006 = pd.read_csv("Pandas_Results2006.csv", index_col = 0) 
data2009 = pd.read_csv("Pandas_Results2009.csv", index_col = 0)
```

```python
fusion = pd.concat([data2006, data2009], ignore_index = True)
```

```python
fusion["name"].value_counts()
```

Fazit: 
- Das Katholische Gymnasium Zürich ist falsch angeschrieben
- Unterstras und Freies Gymnasium kommen nur je einmal vor

```python
fusion[fusion["jahr"] == 2006].sort_values(by="zufr", ascending = False)["zufr"].plot(kind="bar", x="name")
```

### Entwicklung einer einzelnen Schule (alle Werte)

```python
fusion[fusion["name"] == "Glattal"].plot(x="jahr")
```

```python
fusion.groupby("name").plot(x="jahr")
```

```python
fusion.groupby("name")["zufr"].mean().plot(kind = "bar")
```

```python
fusion.groupby("name")["konf"].mean().plot(kind = "bar")
```

```python
fusion.plot(kind='scatter', x='zufr', y='vorb_p')
```

Fazit: Es gibt eine Zusammenhang zwischen allg. Zufreidenheitund Erfolg bei der Persönloichkeitsbildung

```python
fusion.plot(kind='scatter', x='vorb_f', y='vorb_p')
```

Eher weniger stark ist der Zusammenhang zwischen persönlicher und fachlicher Vorbereitung aufs Berufsleben > welche Schulen sind im einen stark und im anderen schwach?

So können wir die Plots nach Jahren auseinander nehmen:

```python
fusion.groupby("jahr").plot(kind='scatter', x='zufr', y='vorb_p')
```
