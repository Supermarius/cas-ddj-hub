Dieser Loop ist fast identisch mit dem von 2015, ich musste bloss die regex leicht modifizieren und die locs der gesuchten Werte um eine Zeile anpassen.

```python
import pandas as pd
import os #Files in listen abgreifen
import re # Regex brauchen wir, um die Namen aus den Filenamen rauszugreifen
```

```python
file_namen = os.listdir('Lieferung_SEB-Schulergebnisse_2/Schulberichte/Tabellen_2018')
jahr = 2018
dic_list = []
```

```python
for file in file_namen:
    regex = r"(?<=4 ).*\b(?= Anhang)"
    name = re.search(regex, file, re.IGNORECASE).group()
    if "Gym" in name:
        regex = r".*\b(?= Gym.*)"
        name = re.search(regex, name, re.IGNORECASE).group()
    if "KS " in name:
        regex = r"(?<=KS ).*\b"
        name = re.search(regex, name, re.IGNORECASE).group()
 
    # Jetzt öffnen wir das jeweilige File als Dataframe:
    
    df = pd.read_excel("Lieferung_SEB-Schulergebnisse_2/Schulberichte/Tabellen_2018/" + str(file))
    
    zufr = df.loc[19,"Unnamed: 3"]
    vorb_f = df.loc[21,"Unnamed: 3"]
    vorb_p = df.loc[23,"Unnamed: 3"]
    prob = df.loc[55,"Unnamed: 3"]
    urt = df.loc[57,"Unnamed: 3"]
    zus = df.loc[59,"Unnamed: 3"]
    konf = df.loc[61,"Unnamed: 3"]

    mini_dict = {"jahr": jahr, 
                 "name": name, 
                 "zufr": zufr, 
                 "vorb_f": vorb_f, 
                 "vorb_p":vorb_p, 
                 "prob": prob, 
                 "urt": urt, 
                 "zus": zus, 
                 "konf": konf}
    dic_list.append(mini_dict)    
```

```python
df_final = pd.DataFrame(dic_list)
df_final.to_csv("Pandas_Results201(.csv")
df_final
```
